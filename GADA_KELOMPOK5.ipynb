### Tahap 2: Menentukan Nilai Matriks Perbandingan Berpasangan



> Langkah 2.1 Menghitung Nilai Komponen Alternatif



```python
df_normalisasi = df_kriteria.div(gm_max, axis=1)

df_hasil_normalisasi = pd.concat([df_alternatif, df_normalisasi], axis=1)
display(df_hasil_normalisasi)
```



> Langkah 2.2 Menghitung Matriks Perbandingan Berpasangan



```python
# Ambil nilai normalisasi
df_normalized_values = df_hasil_normalisasi.drop(columns=['Alternatif']).copy()

# --- Hitung Kolom '1/2 x C1' ---
df_normalized_values['1/2 x C1'] = df_normalized_values['C1'] / 2

# --- Hitung Kolom 'SUM C2-C12' ---
df_normalized_values['SUM C2-C12'] = df_normalized_values['C2'] + df_normalized_values['C12']

# --- Hitung Kolom '1/2 x C13' ---
df_normalized_values['1/2 x C13'] = df_normalized_values['C13'] / 2

# --- Hitung Kolom 'Si' ---
df_normalized_values['Si'] = (
    df_normalized_values['1/2 x C1'] +
    df_normalized_values['SUM C2-C12'] +
    df_normalized_values['1/2 x C13']
)

df_tabel = df_normalized_values[['1/2 x C1', 'SUM C2-C12', '1/2 x C13', 'Si']].copy()
df_tabel_final = pd.concat([df_awal[['Alternatif']], df_tabel], axis=1)
display(df_tabel_final)
```



> Langkah 2.3 Menghitung Elemen Matriks Perbandingan Berpasangan



```python
data_si = {
    'Alternatif': ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10'],
    'Si': [2.833702759, 2.514092606, 2.089273192, 1.953755230, 1.932142032, 2.346026476, 2.616096244, 2.133374509, 1.900652775, 2.820945911]
}
df_si = pd.DataFrame(data_si)
```

```python
Si_values = df_si['Si'].values
alternatif_labels = df_si['Alternatif'].values
n = len(Si_values)

epsilon_matrix = np.zeros((n, n))

# --- Perhitungan Matriks Perbandingan Berpasangan (epsilon_ij) ---
for i in range(n):
    for j in range(n):
        Si = Si_values[i]  # Baris (i)
        Sj = Si_values[j]  # Kolom (j)

        # RUMUS SPESIFIK: (1 + Si + Sj) / (1 + Sj + (Si - Sj))

        # Pembilang (Numerator)
        numerator = 1 + Si + Sj

        # Penyebut (Denominator)
        denominator = 1 + Sj + (Si - Sj)

        # Hasil epsilon_ij
        epsilon_matrix[i, j] = numerator / denominator

# --- Konversi ke DataFrame dan Pembulatan ---
df_epsilon = pd.DataFrame(
    epsilon_matrix,
    index=alternatif_labels,  # Baris (i)
    columns=alternatif_labels # Kolom (j)
).round(9)

# Menambahkan nama untuk Index dan Columns
df_epsilon.index.name = 'Alternatif'
display(df_epsilon)
```

### Tahap 3: Menghitung Bobot Yang Disarankan

```python
# 1. Menghitung Hasil Kali (Perkalian tiap baris)
df_epsilon['Hasil Kali'] = df_epsilon.prod(axis=1)

# 2. Menghitung GM Baris (Rata-rata Geometrik)
df_epsilon['GM Baris'] = df_epsilon['Hasil Kali']**(1/n)

# 3. Menghitung Total (Sigma) GM Baris
total_gm_baris = df_epsilon['GM Baris'].sum()

# 4. Menghitung Bobot Alternatif (Normalisasi)
df_epsilon['Bobot Alternatif'] = df_epsilon['GM Baris'] / total_gm_baris

# Total Bobot Alternatif
total_bobot_alternatif = df_epsilon['Bobot Alternatif'].sum()

# Membuat DataFrame Hasil Akhir
df_hasil = df_epsilon[['Hasil Kali', 'GM Baris', 'Bobot Alternatif']].copy()
df_hasil.index.name = 'Alternatif'
df_hasil = df_hasil.reset_index()

# Membuat baris Total
total_row = pd.DataFrame({
    'Alternatif': ['Total'],
    'Hasil Kali': [''],
    'GM Baris': [total_gm_baris],
    'Bobot Alternatif': [total_bobot_alternatif]
})

df_final = pd.concat([df_hasil, total_row], ignore_index=True)

# Menampilkan Hasil Perhitungan
def format_cell(value):
    if isinstance(value, (float, np.float64)):
        if value == total_gm_baris:
            return f'{value:.9f}'
        elif value == total_bobot_alternatif:
            return f'{value:.0f}'
        elif 'GM Baris' in df_final.columns and value in df_final['GM Baris'].values:
            return f'{value:.9f}'
        elif 'Bobot Alternatif' in df_final.columns and value in df_final['Bobot Alternatif'].values:
            return f'{value:.12f}'
        elif 'Hasil Kali' in df_final.columns and value in df_final['Hasil Kali'].values:
             return f'{value:.8f}'
    return str(value)

df_final_formatted = df_final.applymap(format_cell)

display(df_final_formatted)
```



### Tahap 4: Menghitung Bobot Simulasi dan Kriteria



```python
n = 10
kriteria_labels = [f'C{i}' for i in range(1, n + 1)]
df_epsilon.index = kriteria_labels

df_epsilon_criteria = pd.DataFrame(
    epsilon_matrix,
    index=kriteria_labels,
    columns=kriteria_labels
).round(9)

# 1. Menghitung Hasil Kali (Perkalian tiap kolom)
df_epsilon_criteria['Hasil Kali'] = df_epsilon_criteria.prod(axis=0)

# 2. Menghitung GM Kolom (Rata-rata Geometrik)
df_epsilon_criteria['GM Kolom'] = df_epsilon_criteria['Hasil Kali']**(1/n)

# 3. Menghitung Total (Sigma) GM Kolom
total_gm_kolom = df_epsilon_criteria['GM Kolom'].sum()

# 4. Menghitung Bobot Kriteria (Normalisasi)
df_epsilon_criteria['Bobot kriteria'] = df_epsilon_criteria['GM Kolom'] / total_gm_kolom

# Total Bobot Kriteria
total_bobot_kriteria = df_epsilon_criteria['Bobot kriteria'].sum()

# Membuat DataFrame Hasil Akhir
df_hasil = df_epsilon_criteria[['Hasil Kali', 'GM Kolom', 'Bobot kriteria']].copy()
df_hasil.index.name = 'Kriteria'
df_hasil = df_hasil.reset_index()

# Membuat baris Total
total_row = pd.DataFrame({
    'Kriteria': ['Total'],
    'Hasil Kali': [''],
    'GM Kolom': [total_gm_kolom],
    'Bobot kriteria': [total_bobot_kriteria]
})

df_final = pd.concat([df_hasil, total_row], ignore_index=True)

# Menampilkan Hasil Perhitungan
def format_cell(value):
    if isinstance(value, (float, np.float64)):
        if value == total_gm_kolom:
            return f'{value:.9f}'
        elif value == total_bobot_kriteria:
            return f'{value:.0f}'
        elif 'GM Kolom' in df_final.columns and value in df_final['GM Kolom'].values:
            return f'{value:.9f}'
        elif 'Bobot kriteria' in df_final.columns and value in df_final['Bobot kriteria'].values:
            return f'{value:.12f}'
        elif 'Hasil Kali' in df_final.columns and value in df_final['Hasil Kali'].values:
             return f'{value:.8f}'
    return str(value)

df_final_formatted = df_final.applymap(format_cell)

display(df_final_formatted)
```
